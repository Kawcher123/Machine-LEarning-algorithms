```python
import pandas as pd
df=pd.read_csv('salary.csv')
df
```

```python
x=df.drop(['salary_more_then_100k'],axis='columns')
x
```

```python
y=df.salary_more_then_100k
y
```

```python
x1=pd.get_dummies(x.company)
x2=pd.get_dummies(x.job)
x3=pd.get_dummies(x.degree)
merged=pd.concat([x1,x2,x3],axis='columns')
merged
```

```python
final=merged.drop(['abc pharma','sales executive','masters'],axis='columns')
final
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(final,y,test_size=0.1)
```

```python
X_test
```

```python
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```
